```python
import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-10, 10, 400)
x_sqrt = np.linspace(0, 10, 400)  # sqrt only works for x >= 0

# Define functions
linear = 2 * x + 3
quadratic = x**2
cubic = x**3
sqrt = np.sqrt(x_sqrt)

# Set up the plots
plt.figure(figsize=(12, 8))

# Linear
plt.subplot(2, 2, 1)
plt.plot(x, linear)
plt.title('Linear: f(x) = 2x + 3')
plt.grid(True)

# Quadratic
plt.subplot(2, 2, 2)
plt.plot(x, quadratic)
plt.title('Quadratic: f(x) = x^2')
plt.grid(True)

# Cubic
plt.subplot(2, 2, 3)
plt.plot(x, cubic)
plt.title('Cubic: f(x) = x^3')
plt.grid(True)

# Square Root
plt.subplot(2, 2, 4)
plt.plot(x_sqrt, sqrt)
plt.title('Square Root: f(x) = √x')
plt.grid(True)

plt.tight_layout()
plt.show()
```
